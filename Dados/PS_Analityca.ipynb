```python
data_inscritos = [float(re.sub("[. ]",'',dados_1995['Inscrições em 1 Opção Total'][0])),float(re.sub("[. ]",'',dados_2005['Candidatos Inscritos Total Geral'][0])),float(re.sub("[. ]",'',dados_2015['Seleção para Vagas Novas - Candidatos Inscritos Total Geral'][0])),float(re.sub("[. ]",'',dados_2019['Seleção para Vagas Novas - Candidatos Inscritos Total Geral'][0]))]
```

```python
dados_inscritos = pd.DataFrame(data=data_inscritos,index=['1995','2005','2015','2019'], columns=['Inscritos'])
dados_inscritos
```

```python
data_vagas = [float(dados_1995['Vagas Oferecidas Total'][0]),float(re.sub("[. ]",'',dados_2005['Vagas Oferecidas Total Geral'][0])),float(re.sub("[. ]",'',dados_2015['Seleção para Vagas Novas -  Vagas Oferecidas Total Geral'][0])),float(re.sub("[. ]",'',dados_2019['Seleção para Vagas Novas - Vagas Oferecidas Total Geral'][0]))]
data_vagas
```

```python
dados_vagas = pd.DataFrame(data=data_vagas,index=['1995','2005','2015','2019'], columns=['Vagas'])
dados_vagas
```

```python
dados = pd.concat([dados_inscritos,dados_vagas],axis=1)
```

```python
dados['Vagas'][0] = 610355
```

```python
plt.figure(figsize=(5,5))
grafico = dados.plot.bar(log = False)
grafico.set_title('Vagas e inscritos no Processo Seletivo de IES')
grafico.set_xlabel('Anos')
grafico.set_ylabel('inscrições e vagas')
plt.xticks(rotation = 0)
plt.show()
```

```python
dados_trabalho
```

```python
grafico_trabalho=dados_trabalho.plot.bar(color='lightgreen')
plt.xticks(rotation=0)
#plt.grid(True)
grafico_trabalho.set_xlabel('Tempo')
grafico_trabalho.set_ylabel('Porcentagem')
grafico_trabalho.set_title('Diferença em porcentagem do ano 2002 a 2015')
rects = grafico_trabalho.patches
labels = round(dados_trabalho['Diferença em porcentagem'],2)
for rect, label in zip(rects,labels):
  h = rect.get_height()
  grafico_trabalho.text(rect.get_x()+rect.get_width()/2,h-2,label,ha='center',va='bottom',color='darkgreen')
```
